```python
import speech_recognition as sr
import pyttsx3

# Inicializa el reconocedor
recognizer = sr.Recognizer()

# Inicializa el motor TTS (Text to Speech)
engine = pyttsx3.init()

# Define la función para reconocer voz
def recognize_speech():
    with sr.Microphone() as source:
        print("Di algo...")
        recognizer.adjust_for_ambient_noise(source)  # Ajusta para el ruido ambiental
        audio = recognizer.listen(source)  # Escucha el audio

    try:
        # Utiliza el reconocimiento de voz de Google
        text = recognizer.recognize_google(audio, language="es-ES")  # Cambia el idioma según sea necesario
        print("Has dicho:", text)
        return text.lower()  # Devuelve el texto en minúsculas para facilitar la comparación
    except sr.UnknownValueError:
        print("No se pudo entender lo que dijiste")
        return ""
    except sr.RequestError as e:
        print("Error al solicitar resultados; {0}".format(e))
        return ""

# Define la función para que el programa hable
def speak(text):
    engine.say(text)
    engine.runAndWait()

# Función principal
def main():
    while True:
        # Espera a que el usuario diga "hola"
        while True:
            text = recognize_speech()
            if "hola" in text:
                break

        # Responde con "hola"
        speak("Hola")

        # Espera a que el usuario responda "ola"
        while True:
            text = recognize_speech()
            if "ola" in text:
                break

        # Responde con "Hola Alex"
        speak("Hola Alex")

        # Salir del bucle
        break

# Llama a la función principal
main()
```
